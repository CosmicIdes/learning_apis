# Learning APIs with Rick and Morty API

https://rickandmortyapi.com/api

Info bit gives information on count, pages, what the next page is, and what the previous page is.

https://rickandmortyapi.com/api/character

```python
import requests

import pandas as pd

baseurl = 'https://rickandmortyapi.com/api/'

endpoint = 'character'

```

```python
def main_request(baseurl, endpoint, x):
    r = requests.get(baseurl + endpoint + f'?page={x}')
    return r.json()
```

```python
def get_pages(response):
    return response['info']['pages']
```

```python
def parse_json(response):
    for item in response['results']:
        charlist = []
        char = {
            'id': item['id'],
            'name': item['name'], 
            'no_ep': len(item['episode'])
        }

        charlist.append(char)
    return charlist
```

```python
mainlist = []

data = main_request(baseurl, endpoint, 2)

for x in range(1,get_pages(data)+1):
    print(x)
    mainlist.extend(parse_json(main_request(baseurl, endpoint, x)))

df_rickandmorty = pd.DataFrame(mainlist)

df_rickandmorty.to_csv('charlist.csv', index=False)
```
